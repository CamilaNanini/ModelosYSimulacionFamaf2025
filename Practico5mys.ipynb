# Ejercicio 2

a) Distribución Pareto:
$$
F(x)=\int_{1}^{x} at^{-(a+1)} \ dt\\
= a ⋅ \int_{1}^{x} t^{-(a+1)} \ dt \\
= a ⋅ \frac {t^{n+1}}{n+1}\Big|_1^x \ con \ n=-(a+1)
= a ⋅ [\frac {x^{n+1}}{n+1} -\frac {1^{n+1}}{n+1}] \ con \ n=-(a+1) \\
= a ⋅ \frac {x^{n+1} - 1}{n+1} \ con \ n=-(a+1) \\
(el \ enunciado \ menciona \ que \ a>0 \ asi \ que \ no \ hay \ problema)\\
= a ⋅ \frac {x^{-(a+1)+1} - 1}{-(a+1)+1}\\
= a ⋅ \frac {x^{-a} - 1}{-a} = - (x^{-a} - 1) = 1-x^{-a}
$$
Calcular inversa:
$$
U = 1-x^{-a} ⇒ 1-U = x^{-a} ⇒ (1-U)^{-\frac{1}{a}} = x
$$


b)  Distribución Erlang
Primero recordar que la acumulada de una Gamma es:
$$
Γ(α,β)(x)=\frac{1}{Γ(α)}β^{-α}x^{α-1}e^{-\frac{x}{β}} \\
Si \ α∈N \ entonces:\\
Γ(α,β)(x)=\frac{1}{(α-1)!}β^{-α}x^{α-1}e^{-\frac{x}{β}}=\frac{x^{α-1}e^{-\frac{x}{β}}}{(α-1)!β^{α}}\\
Si \ tomamos \ Γ(α,β)(x) \ con \ α=k \ y \ β=μ \ tenemos \ una \ Erlang.
$$
Y en el teorico vimos que la suma de N v.a con distribución Exponencial y mutuamente independientes genera una distribución Gamma(N,1/λ)

Por lo que tenemos que si $X_i~Exponencial(λ=\frac{1}{μ})$ entonces:
$$
X=∑_{i=1}^{k}X_i \sim Erlang(k,μ)
$$

c) Distribución Weibull
$$
F(x)=\int_{0}^{x} \frac{β}{λ}⋅(\frac{t}{λ})^{β-1} ⋅e^{-(\frac{t}{λ})^{β}} \ dt\\
=  \frac{β}{λ} \int_{0}^{x}(\frac{t}{λ})^{β-1} \ dt \int_{0}^{x} e^{-(\frac{t}{λ})^{β}} \ dt\\
(Confiar \ en \ lo \ que \ wiki \ dice)\\
F(x)= 1 - e^{-(\frac{x}{λ})^β}\\
Luego \ si \ vemos \ la \ inversa:\\
U =  1 - e^{-(\frac{x}{λ})^β} ⇒ U-1 =  - e^{-(\frac{x}{λ})^β} \\
⇒ 1-U = e^{-(\frac{x}{λ})^β} ⇒ 1-U = \frac{1}{e^{(\frac{x}{λ})^β}} \\
⇒ \frac{1}{1-U} = e^{(\frac{x}{λ})^β} ⇒ (\frac{x}{λ})^β = ln(\frac{1}{1-U}) \\
⇒ \frac{x}{λ} = \sqrt[β]{ln(\frac{1}{1-U})} ⇒ x = λ \sqrt[β]{ln(\frac{1}{1-U})}
$$

Las esperanzas de cada una:
- Pareto: $\frac{α}{α-1}$si α>1 sino no existe
- Erlang: kμ
- Weibull: $λ⋅Γ(1+\frac{1}{β})$

```python
import random
import math

def generar_pareto(a):
    u = random.random()
    x = (1 - u)**(-1/a)
    return x

def generar_weibull(beta, lamda):
    u = random.random()
    y = math.log(1/(1-u))
    x = lamda * (y) ** (1/beta)
    return x

def exponencial(lamda):
  U = 1-random.random()
  return -math.log(1-U)/lamda

def generar_erlang(k, mu):
    x = sum(exponencial(1/mu) for _ in range(k))
    return x

a = 2
mu = 2
k = 2
lamda = 1
beta = 2
n_repeticiones = 10000

media_pareto = sum(generar_pareto(a) for _ in range(n_repeticiones)) / n_repeticiones
media_weibull = sum(generar_weibull(beta, lamda) for _ in range(n_repeticiones)) / n_repeticiones
media_erlang = sum(generar_erlang(k, mu) for _ in range(n_repeticiones)) / n_repeticiones

print(f"Media de Pareto real:",a/(a-1))
print(f"Media de Pareto estimada: {media_pareto}")

print(f"\nMedia de Weibull real:",lamda * math.gamma(1+1/beta))
print(f"Media de Weibull estimada: {media_weibull}")

print(f"\nMedia de Erlang real:",k*mu)
print(f"Media de Erlang estimada: {media_erlang}")
```

#Ejercicio 7

f(x) = 1/x si 1 ≤ x ≤ e , 0 en otro caso

Por lo que vamos a calcular la acumulada e inversa.
$$
Para \ 1≤x≤e\\
F(X) = \int_{1}^{x} \frac{1}{t} dt\\
= ln(t)\Big|_1^x\\
...\\
= ln(x) \ para \ 1≤x≤e
$$
Luego la inversa:
$$
U = ln(x) ⇒ x = e^{U}
$$
Ahora para el método de aceptación y rechazo:

- 𝑥∈[1,𝑒]
- 𝑓(𝑥)=1/𝑥
- 𝑔(𝑥)=Uniforme(1,𝑒) ⇒ 𝑔(𝑥)=1/(𝑒−1)
- 𝑐 tal que 𝑓(𝑥) / 𝑔(𝑥) ≤ 𝑐 ⇒ 1/𝑥 ≤ 𝑐 / (𝑒−1). La cota máxima de 1/x en [1,e] es 1, por lo que c=e-1

Luego en el algoritmo:

U ≤ 𝑓(Y) / c . 𝑔(Y) ⇒ U ≤ (1/Y) / (e-1) . (1/(e-1)) ⇒ U ≤ (1/Y)

Cálculo de la esperanza real:
$$
E[X]= ∫_{1}^{e} t \frac{1}{t} dt
=∫_{1}^{e} 1 dt = t\Big|_1^e =e-1 ≈ 1.71828
$$

Probabilidad de que X ≤ 2
$$
P(X≤2)=F(2)=ln(2)≈0.6931
$$

```python
import random
import math

# Método de la Transformada Inversa
def X_TI():
    U = random.random()
    return math.e ** U

# Método de Aceptación y Rechazo
def X_AR():
    while True:
        Y = random.uniform(1, math.e)
        U = random.random()
        if U < 1 / Y:
            return Y


n_sim = 10000
muestras_TI = [X_TI() for _ in range(n_sim)]
muestras_AR = [X_AR() for _ in range(n_sim)]

esperanza_TI = sum(muestras_TI) / n_sim
esperanza_AR = sum(muestras_AR) / n_sim
esperanza_real = math.e - 1

prob_TI = sum(x <= 2 for x in muestras_TI) / n_sim
prob_AR = sum(x <= 2 for x in muestras_AR) / n_sim
prob_real = math.log(2)

print("ESPERANZAS")
print(f"Real: {esperanza_real:.5f}")
print(f"TI: {esperanza_TI:.5f}")
print(f"AR: {esperanza_AR:.5f}")
print()
print("P(X ≤ 2)")
print(f"Real: {prob_real:.5f}")
print(f"TI: {prob_TI:.5f}")
print(f"AR: {prob_AR:.5f}")
```

$$

$$

$$

```python
import random
import math

def eventosPoisson(lamda,T):
  t = 0
  NT = 0
  Eventos = []
  while t < T:
    U = 1 - random.random()
    t += - math.log(U) / lamda
    if t <= T:
      NT += 1
      Eventos.append(t)
  return NT, Eventos
```

```python
import random
import math

#a
def padelgazamiento(lamda_t, lamda_max,T):
  '''
  lamda_t : funcion de intensidad
  lamda_max : valor máximo que acota la intensidad
  T : Intervalo de tiempo en el que se simulan los eventos
  '''
  NT = 0
  Eventos = []
  U = 1 - random.random()
  t = -math.log(U) / lamda_max
  while t <= T:
    V = random.random()
    if V < lamda_t(t) / lamda_max:
      NT += 1
      Eventos.append(t)
    t += -math.log(1 - random.random()) / lamda_max
  return NT, Eventos

def lambda1(t):
    return 3 + 4 / (t + 1)

def lambda2(t):
    return (t - 2)**2 - 5*t + 17

def lambda3(t):
    if 2 <= t <= 3:
        return t / 2 - 1
    elif 3 < t <= 6:
        return 1 - t / 6
    else:
        return 0

NT1, eventos1 = padelgazamiento(lambda1, 7, 3)
print("Caso i):", NT1, "eventos:", eventos1)
NT2, eventos2 = padelgazamiento(lambda2, 21, 5)
print("Caso ii):", NT2, "eventos:", eventos2)
NT3, eventos3 = padelgazamiento(lambda3, 0.5, 6)
print("Caso iii):", NT3, "eventos:", eventos3)

def estimar_esperanza(lambda_t, lambda_max, T, n=10000):
    total_eventos = 0
    for _ in range(n):
        NT, _ = padelgazamiento(lambda_t, lambda_max, T)
        total_eventos += NT
    return total_eventos / n

esp1 = estimar_esperanza(lambda1, 7, 3)
print("Esperanza simulada (caso i):", esp1)

esp2 = estimar_esperanza(lambda2, 21, 5)
print("Esperanza simulada (caso ii):", esp2)

esp3 = estimar_esperanza(lambda3, 0.5, 6)
print("Esperanza simulada (caso iii):", esp3)
```

Sub-intervalos:

i) $λ(t)=3+\frac{4}{t+1}$ con 0≤t≤3 y vamos a tomar los intervalos [0,1),[1,2),[2,3), analizando cada intervalo tenemos los siguientes $λ_{maximos}$:

- [0,1) ⇒ λ(0) = 7
- [1,2) ⇒ λ(1) = 5
- [2,3) ⇒ λ(2) = $\frac{13}{3}$

ii) $λ(t)=(t-2)^2-5t+17$ con 0≤t≤5 y vamos a tomar los intervalos [0,1),[1,2),[2,3),[3,4),[4,5) analizando cada intervalo tenemos los siguientes $λ_{maximos}$:

- [0,1) ⇒ λ(0) = 21
- [1,2) ⇒ λ(0) = 13
- [2,3) ⇒ λ(0) = 7
- [3,4) ⇒ λ(0) = 3
- [4,5) ⇒ λ(0) = 1

iii) $λ(t)=\frac{t}{2}-1$ si 2≤t≤3 y $λ(t)=1-\frac{t}{6}$ si 3< t ≤6 y vamos a tomar los intervalos [2,3),[3,4),[4,5),[5,6) analizando cada intervalo tenemos los siguientes $λ_{maximos}$:

- [2,3) ⇒ λ(x) = 0.5 (con x→3)
- [3,4) ⇒ λ(3) = 0.5
- [4,5) ⇒ λ(4) = $\frac{1}{3}$
- [5,6) ⇒ λ(5) = $\frac{1}{6}$

```python
#b
import random
import math

def Poisson_adelgazamiento_mejorado(interv, lamdas, T, lamda_t):
  '''
  lamda_t : funcion de intensidad
  lamda_max : valor máximo que acota la intensidad
  T : Intervalo de tiempo en el que se simulan los eventos
  '''
  j = 0
  t = -math.log(random.random()) / lamdas[j]
  NT = 0
  Eventos = []
  while t <= T:
    if t <= interv[j]:
      V = random.random()
      if V < lamda_t(t) / lamdas[j]:
        NT += 1
        Eventos.append(t)
      t += -math.log(1 - random.random()) / lamdas[j]
    else: #t > interv[j]
      t = interv[j] + (t - interv[j]) * lamdas[j] / lamdas[j + 1]
      j += 1
  return NT, Eventos


def lambda1(t):
    return 3 + 4 / (t + 1)
interv1 = [1, 2, 3]
lamdas1 = [7.0, 6.0, 5.5]

def lambda2(t):
    return (t - 2)**2 - 5*t + 17
interv2 = [1, 2, 3, 4, 5]
lamdas2 = [21, 13, 7, 3, 1]

def lambda3(t):
    if 2 <= t <= 3:
        return t / 2 - 1
    elif 3 < t <= 6:
        return 1 - t / 6
    else:
        return 0
interv3 = [3,4,5,6]
lamdas3 = [0.5, 0.5, 1/3, 1/6]

def estimar_esperanza(lamda_t, lamdas, interv, T, n=10000):
    total_eventos = 0
    for _ in range(n):
        NT, _ = Poisson_adelgazamiento_mejorado(interv, lamdas, T, lamda_t)
        total_eventos += NT
    return total_eventos / n

esp1 = estimar_esperanza(lambda1,lamdas1,interv1,3)
print("Esperanza simulada (caso i):",esp1)
esp2 = estimar_esperanza(lambda2,lamdas2,interv2,5)
print("Esperanza simulada (caso ii):",esp2)
esp3 = estimar_esperanza(lambda3,lamdas3,interv3,6)
print("Esperanza simulada (caso iii):",esp3)
```
