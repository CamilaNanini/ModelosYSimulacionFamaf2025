a) Se trabajará con la siguiente hipótesis nula:

$H_0$: los datos provienen de una distribución exponencial de paramétro λ=0.5

Para ello se utilizará el test de Kolmogorov-Smirnov en donde en primera instancia calcularemos el estadístico:$D = \sup_{x∈R}|F_e(x)-F(x)|$ en donde $F_e$ representa la función empírica del problema (o sea los datos que se observaron) y $F$ es la función de la distribución supusimos en nuestra $H_0$. Posteriormente se realizaran varias simulaciones generando muestras con nuestra distribución supuesta y calcularemos reiteradas veces el estadístico obtenido en estas simulaciones con el estadístico observado contando aquellas veces en el que el simulado fue mayor que el observado y posteriormente dividiendo este conteo por la cantidad de muestras simuladas. Este procedimiento nos dará una estimación del p-valor el cual nos indicará si debemos rechazar nuestra $H_0$ o si no hay evidencia suficiente para rechazar la misma.

Para poder hacer esta útlima comparación vamos a utilizar un cierto alpha y diremos que:

si p_val < alpha: Se rechaza H_0 y los datos no provienen de la distribución mencionada
Sino: No hay suficiente evidencia para rechazar la H_0 por lo que los datos podrían provenir de la distribución pensada.

En este caso tenemos un parámetro desconocido el cual es p, el mismo lo estimaremos de la siguiente forma: $p'=\frac{1}{n}∑_{i=1}^{5}x_i*c_i$.
Entonces $p'= 0⋅38+1⋅144+2⋅342+3⋅287+4⋅164+5⋅25= 2470/5000$ entonces p'= 0.494
Al igual que en el punto anterior se calculara el estadistico qeu proviene de los datos dados para posteriormente hacer varias muestras simuladas y comoparar los estadísticos que se obtuvieron, comparandolos entre si y contando aquellos que son mayores al observado, para posteriormente dividirlo por la cantidad de simulaciones hecas y así obterner una estimación de lo que es el p-valor. Para posteriormente contrastar este resultado obtenido con el alpha considerado.

b) Calcular el valor del estadistico.

El estadístico $T=∑_{i=1}^{n}\frac{(N_i-np_i)^{2}}{np_i}$ donde en este caso $p_i = \binom{5}{i} \ 0.494^{i} \ (1-0.494)^{5-i} \ ∀i∈{0,1,2,3,4,5}$ así que: $T=\frac{(38-5000 \cdot p0)^{2}}{5000 \cdot p0}+\frac{(144-5000 \cdot p1)^{2}}{5000 \cdot p1}+\frac{(684-5000 \cdot p2)^{2}}{5000 \cdot p2}+\frac{(861-5000 \cdot p3)^{2}}{5000 \cdot p3}+\frac{(656-5000 \cdot p4)^{2}}{5000 \cdot p4}+\frac{(125-5000 \cdot p5)^{2}}{5000 \cdot p5}≈\fbox{8.1616}$


```python
import numpy as np
from collections import Counter
from scipy.stats import binom
def ejercicio3():
  def Binomial(n,p):
    c = p / (1 - p)
    prob = (1 - p) ** n
    F = prob; i=0
    U = random()
    while U >= F:
      prob *= c * (n-i) / (i+1)
      F += prob
      i += 1
    return i

  def contar_frecuencias_counter(observaciones, valores_posibles):
      conteos = Counter(observaciones)
      return [conteos[val] for val in valores_posibles]

  def chi_squared_statistic(observed, expected):
      total = 0
      for i in range(len(observed)):
        o = observed[i]
        e = expected[i]
        total += (o - e) ** 2 / e
      return total

  def simulate_p_value(n, observed_stat, p, num_simulations=1000):
      count = 0
      valores_posibles = [0, 1, 2, 3, 4, 5]
      for _ in range(num_simulations):
          sim = [Binomial(5,p) for i in range(n)]
          simulated_observed = contar_frecuencias_counter(sim, valores_posibles)
          expected_sim = []
          for k in valores_posibles:
            binom_pmf = (math.comb(5, k) * (p**k) * ((1 - p)**(5 - k)))
            expected_sim.append(n * binom_pmf)
          simulated_stat = chi_squared_statistic(simulated_observed, expected_sim)
          if simulated_stat >= observed_stat:
              count += 1
      return count / num_simulations


  observed = [38, 144, 342, 287, 164, 25]
  valores_posibles = [0, 1, 2, 3, 4, 5]
  n = sum(observed)
  total_caras = 0
  for i in range(len(valores_posibles)):
    total_caras += i * observed[i]
  p_hat = total_caras / (5 * n)
  expected = []
  for k in valores_posibles:
      binom_pmf = (math.comb(5, k) * (p_hat**k) * ((1 - p_hat)**(5 - k)))
      expected.append(n * binom_pmf)

  observed_stat = chi_squared_statistic(observed, expected)

  N = 1000
  p_val = simulate_p_value(n, observed_stat, p_hat, N)
  alpha = 0.05

  print(f"p_hat = {p_hat:.4f}")
  print(f"Estadístico chi-cuadrado observado: {observed_stat:.4f}")
  print(f"p-valor estimado por simulación ({N} repeticiones): {p_val:.4f}")
  print("Se rechaza H_0" if p_val < alpha else "No hay evidencia para rechazar H_0")
```

En este ejecicio se debe de hacer la aproximacion a la integral por medio de monte carlo, para esto vamos a tener en cuneta el cambio de variable $y=\frac{x-a}{b-a}$ y $dy=\frac{1}{b-a}dx$ pues la integral ocurre en el intervalo [2,3]

```python
from random import random
import math
from scipy.stats import norm

def ejercicio4():
  def funcion(x):
      return math.exp(-x) * (1 - x**4)

  def monte_carlo_integral(g, a, b, z_alfa_2, L):
      d = L / (2 * z_alfa_2)
      n = 1
      x = a + (b - a) * random()
      val = g(x)
      media = val
      Scuad = 0

      while n <= 100 or math.sqrt(Scuad / n) > d:
          n += 1
          x = a + (b - a) * random()
          val = g(x)
          media_ant = media
          media = media_ant + (val - media_ant) / n
          Scuad = Scuad * (1 - 1 / (n - 1)) + n * (media - media_ant)**2

          if n in [1000, 5000, 7000]:
              print(f"\nNúmero de simulaciones: {n}")
              print(f"Estimación final: {round((b-a)*media, 4)}")
              print(f"Intervalo de confianza: ({round(media - 1.96 * math.sqrt(Scuad)/ math.sqrt(n), 4)}, {round(media + 1.96 * math.sqrt(Scuad) / math.sqrt(n), 4)})")

      I_estimado = (b - a) * media
      return n, I_estimado, Scuad

  a = 2
  b = 3
  L = 2 * 0.001
  cf = 0.95
  z_alfa_2 = norm.ppf(1 - (1 - cf)/2)

  print(f"Valor de z_(alfa/2): {z_alfa_2}")
  print("Valor real aproximado de la integral: -3.08446150031")

  n, estimacion, _ = monte_carlo_integral(funcion, a, b, z_alfa_2, L)
  print(f"\nEstimación final de la integral: {round(estimacion, 4)}")
  print(f"Número total de simulaciones: {n}")
```
